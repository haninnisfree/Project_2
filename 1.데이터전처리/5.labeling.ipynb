```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/5. 이자보상배율 완료.csv",encoding='cp949')
```

상장폐지

```python
df.reset_index(inplace=True)
df_label = df[['index','회사명', '거래소코드', '회계년도', '상장일', '상장폐지일', '소속코드']]
df_label = df_label[df_label['소속코드'] == 4].groupby('회사명')['index'].apply(list)
df_label = pd.DataFrame(df_label)

for j in df_label.index:
    for i in df.index:
        if df_label["index"][j][-1] == i:
            df.loc[i,'label'] = 1
df['label'] = df["label"].fillna(0)
```

관리종목

```python
df_20121 = pd.read_csv("../data/2012년 관리종목 지정 종목 1.csv", encoding='cp949')
df_20122 = pd.read_csv("../data/2012년 관리종목 지정 종목 2.csv", encoding='cp949')
df_2012 = pd.concat([df_20121,df_20122])

df_2012.columns =['번호', '시간', '회사명', '거래소코드', '공시제목', '제출인']
df_all_2012 = df_2012["거래소코드"].unique().tolist()

len(df_all_2012)

all_list = []
for i in df_all_2012:
    a = len(df[df["거래소코드"]== i])
    if a >= 1 and  df[df["거래소코드"]== i]["label"].sum()<1:
        b = df_2012[df_2012["거래소코드"]==i]
        all_list.append(b)

all_df = pd.concat(all_list)    

list_2012 = all_df[all_df["공시제목"].str.contains("지정")]["거래소코드"].unique().tolist()
index_list_2012 = []
del_list_2012=[]
for i in list_2012:
    a = df[df["거래소코드"]== i]
    num = a[a["회계년도"]=="2011-12-01"].index
    num_list = a[a["회계년도"]!="2011-12-01"].index
    index_list_2012.append(num)
    del_list_2012.append(num_list)

for i in index_list_2012:
    df.loc[i,"label"] = 1
for c in del_list_2012:
    for j in c:
        df.drop(j, inplace=True)
```
